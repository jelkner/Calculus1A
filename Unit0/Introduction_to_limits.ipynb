```python
def f(x):
    return (3 - 5 * x + x ** 2 + x ** 3) ** 0.5 / (x - 1)
```

# Example
Approaching x from left

```python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
x = numpy.linspace(0, 1, 1000, endpoint=False)
y = f(x)
plt.plot(x, y)
```
